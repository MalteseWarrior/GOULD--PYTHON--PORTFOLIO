The purpose of this code is to utilize spaCy's EntityRuler to create and apply rules for extracting entites representing companies and stock symbols from a dataframe


```python
## Import Necessary Libraries
import pandas as pd
import spacy
nlp = spacy.load("en_core_web_sm")
from spacy.pipeline import EntityRuler

## Task 2
# Load the datafile as a dataframe 
df = pd.read_csv('stocks-1.tsv', sep='\t')


## Task 3
# Define Patterns based on the stock symbol and company name - For loop in case of dataframe additons
patterns = []
for i in range(len(df)):
    # Company names have spaces and need to be tokenized correctly - This took a while...
    patterns.append({'label': 'Stock Symbol', 'pattern': [{'TEXT': df['Symbol'][i]}]})
    patterns.append({'label': 'Company Name', 'pattern': [{'TEXT': token.text} for token in nlp.make_doc(df['CompanyName'][i])]})
                    
## Task 4
# Create the Ruler and add the patterns, Data Science Lab suggested to use the before="ner" option to avoid conflicts with the NER component
ruler = nlp.add_pipe("entity_ruler", before="ner")
ruler.add_patterns(patterns)


## Task 5
# Define the texts to be processed by the Ruler
# Opted for list format for easier iteration and printing
texts = [
    "Helmerich & Payne (HP) saw its stock rise by 1.5%, fueled by optimistic forecasts in the Energy Equipment & Services sector."
    "In contrast, Check-Cap (CHEK) faced a decline of 2.3% following its announcement of increased costs related to supply chain disruptions."
    "Meanwhile, Vallon Pharmaceuticals (VLON) gained 0.8% after strong quarterly earnings, outperforming its peers in the Biotechnology space."
    "Sequans Communications (SQNS) also recorded a modest increase of 0.5%, reflecting investors' confidence in its ability to navigate challenges in the Semiconductors & Semiconductor Equipment industry.",

    "Aemetis (AMTX) saw its stock rise by 1.5%, fueled by optimistic forecasts in the Oil, Gas & Consumable Fuels sector."
    "In contrast, Ferro Corporation (FOE) faced a decline of 2.3% following its announcement of increased costs related to supply chain disruptions."
    "Meanwhile, RingCentral (RNG) gained 0.8% after strong quarterly earnings, outperforming its peers in the Software space. ACI Worldwide (ACIW)"
    "also recorded a modest increase of 0.5%, reflecting investors' confidence in its ability to navigate challenges in the Software industry.",

    "On a mixed trading day, Par Pacific Holdings (PARR) saw its stock rise by 1.5%, fueled by optimistic forecasts in the Oil, Gas & Consumable Fuels sector."
    "In contrast, Nano Dimension (NNDM) faced a decline of 2.3% following its announcement of increased costs related to supply chain disruptions."
    "Meanwhile, Beyond Meat (BYND) gained 0.8% after strong quarterly earnings, outperforming its peers in the Food Products space."
    "Apollo Investment (AINV) also recorded a modest increase of 0.5%, reflecting investors' confidence in its ability to navigate challenges in the Capital Markets industry."

]

# Process the texts and extract entities using the Ruler
# Iterate through the texts and print the entities found
# Added a counter to track the paragraph number
textnum = 0
for text in texts:
    textnum += 1
    doc = nlp(text)
    print("Paragraph: ", textnum)
    for ent in doc.ents:
        if ent.label_ == "Stock Symbol":
            print(ent.text + " is a " + ent.label_)
        elif ent.label_ == "Company Name":
            print(ent.text + " is a " + ent.label_)
    print("\n")
```
